### 15) Define a function which takes in a a numpy array of shape (length, height, depth) and returns a vector of shape (length\*height*depth, 1)


```python
#your code here
def multiToSingle(ar):
    l1 = len(ar)
    l2 = len(ar[0])
    l3 = len(ar[0][0])
    ar2 = np.zeros(l1*l2*l3)
    count = 0
    for x in ar:
        for y in x:
            for z in y:
                ar2[count] = z
                count+=1
    print(ar2)
    
multiToSingle([[[1,2]
                ,[3,4]],
               [[5,6],
                [7,8]]])
```

### 16) Define a softmax function that calculates the softmax for each row of the input x.

Note:
- The code should work for a row vector and also for matrices of shape (n, m)

**Test Case:**

**Input:**

 [[9 2 5 0 0]

   [7 5 0 0 0]]

Output:

[[9.80897665e-01 8.94462891e-04 1.79657674e-02 1.21052389e-04
  1.21052389e-04]

 [8.78679856e-01 1.18916387e-01 8.01252314e-04 8.01252314e-04
  8.01252314e-04]]

```python
#your code here
def softmax(ar,r,c):
    result = np.zeros((r,c))
    for x in range(0,r):
        sum = np.exp(ar[x]).sum()
        for y in range(0,c):
            result[x][y] = np.exp(ar[x][y])/(sum)
    print(result)
r = int(input("Enter number of rows"))
c = int(input("Enter number of columns"))
ar1 = np.ones((r, c), dtype=int)
for x in range(0,r):
    for y in range(0,c):
        ar1[x,y] = input("Number : ")
softmax(ar1,r,c)
```
