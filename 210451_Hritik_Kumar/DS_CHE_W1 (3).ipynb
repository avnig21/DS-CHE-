### 14) Implement the derivative of the sigmoid function using numpy. 

Note: 
- Read about sigmoid function over the internet.
- Input of the function can be a scalar or a numpy array.

**Test Case:**

**Input:**

 [[9, 2, 5, 0, 0]

   [7, 5, 0, 0, 0]]

Output:

[[0.99987661, 0.88079708, 0.99330715, 0.5, 0.5 ]

[0.99908895, 0.99330715, 0.5,  0.5,  0.5 ]]

```python
#your code here
def sigmoid(a):
    return np.exp(-a)/(1+np.exp(-a))**2
a=np.array([[9,2,5,0,0],[7,5,0,0,0]])
ans=sigmoid(a)
ans=np.array(ans,dtype=float)
ans
```

### 16) Define a softmax function that calculates the softmax for each row of the input x.

Note:
- The code should work for a row vector and also for matrices of shape (n, m)

**Test Case:**

**Input:**

 [[9 2 5 0 0]

   [7 5 0 0 0]]

Output:

[[9.80897665e-01 8.94462891e-04 1.79657674e-02 1.21052389e-04
  1.21052389e-04]

 [8.78679856e-01 1.18916387e-01 8.01252314e-04 8.01252314e-04
  8.01252314e-04]]

```python
#your code here
def softmax(a):
    x=a.shape[0]
    if(a.shape==(x,)):
        #a=a.reshape(x,1)
        return np.exp(a)/sum(np.exp(a))
    b=np.zeros(a.shape)
    for i in range(x):
        b[i,:]=np.exp(a[i,:])/sum(np.exp(a[i,:]))
    return b
```

```python
x=np.array([[9,2,5,0,0],[7,5,0,0,0]])
b=softmax(x)
l=x.shape[0]

print(b)
```
