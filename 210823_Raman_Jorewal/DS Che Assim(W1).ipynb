### 16) Define a softmax function that calculates the softmax for each row of the input x.

Note:
- The code should work for a row vector and also for matrices of shape (n, m)

**Test Case:**

**Input:**

 [[9 2 5 0 0]

   [7 5 0 0 0]]

Output:

[[9.80897665e-01 8.94462891e-04 1.79657674e-02 1.21052389e-04
  1.21052389e-04]

 [8.78679856e-01 1.18916387e-01 8.01252314e-04 8.01252314e-04
  8.01252314e-04]]

```python
# y = np.exp(x)/np.sum(np.exp(x))
def smx(h):
    y=h.shape[0]
    if(h.shape==(y,)):
        return np.exp(h)/sum(np.exp(h))
    k=np.zeros(h.shape)
    for i in range(y):
        k[i,:]=np.exp(h[i,:])/sum(np.exp(h[i,:]))
    return k

y = np.array([[9, 2, 5, 0, 0], [7, 5, 0, 0, 0]])
k = smx(y)
p = y.shape[0]
print(k)
```
